```python
import pandas as pd
import numpy as np
```

```python
# YEAR 2000
df = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df.info()
df.head()
```

```python
# YEAR 2001
df2 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df2.info()
df2.head()
```

```python
# Concatenate two dataframes 
df3 = pd.concat([df, df2], axis = 0)
df3.info()
df3.head()
```

**How many movies had at least some valid financial information (values > 0 for budget OR revenue)?**

```python
# df2=df.loc[(df['Discount'] >= 1000) & (df['Discount'] <= 2000)]
df4 = df3.loc[df3['budget'] > 0]
df4.info()
df4.head()
```

```python
df5 = df3.loc[df3['revenue'] > 0]
df5.info()
df5.head()
```

```python
## Excludes movies with 0's for budget and revenue
df6=df3.loc[(df3['revenue'] > 0) & (df3['budget'] >0)]
df6.info()
df6.head()
```

**Conclusion:**

* df4 (budget) has 545 entries
* df5 (revenue) has 446 entries
* df6 (both columns) has 354 entries 
* Total = 991 movies

** 2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?**

```python
## 13 movies with financial info that have G rating
# Removes scientific notation
pd.set_option('display.float_format', lambda x: f'{x:.3f}')

df7 = df6.loc[df6['certification'] == 'G']
df7_mean = df7[["revenue","budget"]].mean()
print(df7_mean)
print('\n')
df7.info()
```

```python
# 30 movies with PG rating
df8 = df6.loc[df6['certification'] == 'PG']
df8_mean = df8[["revenue","budget"]].mean()
print(df8_mean)
print('\n')
df8.info()
```

```python
# 117 movies with financial info and pg13 rating
df9 = df6.loc[df6['certification'] == 'PG-13']
df9_mean = df9[["revenue","budget"]].mean()
print(df9_mean)
print('\n')
df9.info()
```

```python
# 142 movies with financial info and R rating
df10 = df6.loc[df6['certification'] == 'R']
df10_mean = df10[["revenue","budget"]].mean()
print(df10_mean)
print('\n')
df10.info()
```
