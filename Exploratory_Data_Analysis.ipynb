# EDA

## Import and Load

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# YEAR 2000
df = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df.info()
df.head()
```

```python
# YEAR 2001
df2 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df2.info()
df2.head()
```

**Concatenating Dataframes**

```python
# Concatenate two dataframes 
df3 = pd.concat([df, df2], axis = 0)
df3.info()
df3.head()
```

## Questions

```python
# Filtering for movies with a budget >0 ... 545 movies
df4 = df3.loc[df3['budget'] > 0]
df4.info()
df4.head()
```

```python
# Remove scientific notation
pd.set_option('display.float_format', lambda x: f'{x:.3f}')
```

**Histplot for Movie Budgets**

```python
# Histplot for movies with budget >0 
sns.color_palette('pastel')
ax = sns.histplot(data = df4, x = 'budget', bins = 50)

ax.set(title = 'Movie Budgets', xlabel = 'Budget')

median = df4['budget'].median()

ax.axvline(median, color = 'black', ls = '--',
           label = f'Median Budgets = ${median}')

ax.legend();
```

* The budget distribution is right skewed
* Most budgets occur below the median

```python
# Filerting for movies with a revenue > 0 .. 446 
df5 = df3.loc[df3['revenue'] > 0]
df5.info()
df5.head()
```

**Histplot for Movie Revenues**

```python
# Histplot for movies with revenue >0 

ax = sns.histplot(data = df5, x = 'revenue', bins = 50)

ax.set(title = 'Movie Revenue', xlabel = 'Revenue')

median = df5['revenue'].median()

ax.axvline(median, color = 'black', ls = '--',
           label = f'Median Movie Revenue = ${median}')

ax.legend();
```

* Movie revenues are very right skewed
* Most revenues occur at or below the median

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

**How many movies had at least some financial information**

```python
filter_financial = (df3['budget'] >0) | (df3['revenue'] >0)
filter_financial.value_counts()
```

```python
explore_df = df3[filter_financial].copy()
```

```python
rating_order = ['G', 'PG', 'PG-13', 'R', 'NR']
```

**2. How many movies are there in each of the certification categories**

```python
explore_df.groupby('certification')['certification'].count()
```

```python
ax = sns.countplot(data = explore_df, x = 'certification', 
                   order = rating_order);

ax.set_title('Movie Count by Certification', fontsize = 10, fontweight = 'bold');
ax.set_xlabel('Certification', fontsize = 8, fontweight = 'bold')
ax.set_ylabel('Count', fontsize = 8, fontweight = 'bold');
```

**What is the average revenue per certification category?**

```python
explore_df.groupby('certification')['revenue'].mean()
```

```python
ax = sns.barplot(data = explore_df, x = 'certification', y = 'revenue',
                 order = rating_order);

ax.set_title('Average Revenue by Certification', fontsize = 10, fontweight = 'bold');
ax.set_xlabel('Certification', fontsize = 8, fontweight = 'bold')
ax.set_ylabel('Budget', fontsize = 8, fontweight = 'bold');
```

**What is the average budget per certification category?**

```python
explore_df.groupby('certification')['budget'].mean()
```

```python
ax = sns.barplot(data = explore_df, x = 'certification', y = 'budget', 
            order = rating_order);

ax.set_title('Average Budget by Certification', fontsize = 10, fontweight = 'bold');
ax.set_xlabel('Certification', fontsize = 8, fontweight = 'bold')
ax.set_ylabel('Budget', fontsize = 8, fontweight = 'bold');
```
