# EDA

## Import and Load

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# YEAR 2000
df = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df.info()
df.head()
```

```python
# YEAR 2001
df2 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df2.info()
df2.head()
```

**Concatenating Dataframes**

```python
# Concatenate two dataframes 
df3 = pd.concat([df, df2], axis = 0)
df3.info()
df3.head()
```

## Questions

**How many movies had at least some valid financial information (values > 0 for budget OR revenue)?**

```python
# Movies with a budget
df4 = df3.loc[df3['budget'] > 0]
df4.info()
df4.head()
```

```python
# Remove scientific notation
pd.set_option('display.float_format', lambda x: f'{x:.3f}')
```

**Histplot for Movie Budgets**

```python
# Histplot for movies with budget >0 
sns.color_palette('pastel')
ax = sns.histplot(data = df4, x = 'budget', bins = 50)

ax.set(title = 'Movie Budgets', xlabel = 'Budget')

median = df4['budget'].median()

ax.axvline(median, color = 'black', ls = '--',
           label = f'Median Budgets = ${median}')

ax.legend();
```

* The budget distribution is right skewed
* Most budgets occur below the median

```python
# Movies with a revenue
df5 = df3.loc[df3['revenue'] > 0]
df5.info()
df5.head()
```

**Histplot for Movie Revenues**

```python
# Histplot for movies with revenue >0 

ax = sns.histplot(data = df5, x = 'revenue', bins = 50)

ax.set(title = 'Movie Revenue', xlabel = 'Revenue')

median = df5['revenue'].median()

ax.axvline(median, color = 'black', ls = '--',
           label = f'Median Movie Revenue = ${median}')

ax.legend();
```

* Movie revenues are very right skewed
* Most revenues occur at or below the median

**Conclusion:**

* df4 (budget): 545 entries > 0
* df5 (revenue): 446 entries > 0
* Total = 991 movies with EITHER financial info > 0

## Dropping Budget and Revenue = 0

```python
# Excluding movies with 0's for budget and revenue
df3_new = df3.drop(df3[(df3['budget'] == 0) & (df3['revenue'] == 0)].index)
df3_new.info()
df3_new.head()
```

**How many movies are there in each of the certification categories (G/PG/PG-13/R)?**

```python
## 5 movies with both financial info that have G rating
## mean for revenue and budget
df7 = df3_new.loc[df3_new['certification'] == 'G']
df7_mean = df7[["revenue","budget"]].mean()
print(df7_mean)
print('\n')
df7.info()
```

```python
# 18 movies with PG rating
## mean for revenue and budget
df8 = df3_new.loc[df3_new['certification'] == 'PG']
df8_mean = df8[["revenue","budget"]].mean()
print(df8_mean)
print('\n')
df8.info()
```

```python
# 47 movies with financial info and pg13 rating
## mean for revenue and budget
df9 = df3_new.loc[df3_new['certification'] == 'PG-13']
df9_mean = df9[["revenue","budget"]].mean()
print(df9_mean)
print('\n')
df9.info()
```

```python
# 86 movies with financial info and R rating
## mean for revenue and budget
df10 = df3_new.loc[df3_new['certification'] == 'R']
df10_mean = df10[["revenue","budget"]].mean()
print(df10_mean)
print('\n')
df10.info()
```

**Countplot for Certification**

```python
sns.countplot(data= df3_new, x='certification');
```

* Movies with R rating had the most financial data available. 
* PG-13 movies has the second highest amount

**Average Revenue per Rating**

```python
# Average revenue per certification category
means = df3_new.groupby('certification')['revenue'].mean().sort_values(ascending = False)
means
```

```python
fig, ax = plt.subplots(figsize=(20,10))

ax = sns.barplot(data=df3_new, x='certification', y = 'revenue', order = means.index,
                 errorbar = None);


plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12);
ax.set_title('Average Revenue by Rating', fontsize = 20, fontweight = 'bold');
ax.set_xlabel('Rating', fontsize = 15, fontweight = 'bold')
ax.set_ylabel('Revenue', fontsize = 15, fontweight = 'bold');


ticks = ax.get_yticks();
new_labels = ['${}'.format(int(amt)) for amt in ticks];
ax.set_yticklabels(new_labels);


# Help from Exemplar student project
#https://keithwithdata.medium.com/this-week-i-learnt-how-to-add-to-ticks-on-matplotlib-graphs-f67ccee1f3dc
```
